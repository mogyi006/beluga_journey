```python
# Create sample data
data = {
    'Name': ['John Doe', 'Jane Smith', 'Michael Johnson', 'Emily Davis', 'David Wilson',
             'Sarah Brown', 'James Miller', 'Laura Taylor', 'Daniel Anderson', 'Olivia Thomas'],
    'SSN': ['123-45-6789', '987-65-4321', '456-78-9012', '345-67-8901', '234-56-7890',
            '567-89-0123', '678-90-1234', '789-01-2345', '890-12-3456', '901-23-4567'],
    'Phone Number': ['202-555-0198', '305-555-0172', '212-555-0145', '415-555-0109', '512-555-0190',
                     '617-555-0157', '718-555-0163', '202-555-0137', '213-555-0181', '303-555-0149'],
    'Email': ['johndoe@gmail.com', 'janesmith@hotmail.com', 'michaeljohnson@yahoo.com', 
              'emily.davis@abc_company.com', 'david.wilson@road_company.com', 
              'sarah.brown@gmail.com', 'james.miller@yahoo.com', 
              'laura.taylor@hotmail.com', 'daniel.anderson@ai_company.com', 
              'olivia.thomas@gmail.com']
}

# Convert data to a DataFrame
df = pd.DataFrame(data)

display(df)
```

## Anonymization

With anonymization we can safely store personal data at rest, however the original values will be lost forever.

```python
df_anon = df.copy()

df_anon['anon_id'] = np.random.randint(0, 9999, df_anon.shape[0])
df_anon["Email"] = df_anon["Email"].apply(lambda x: "*******@" + x.split("@")[1])
df_anon['Phone Number'] = df_anon['Phone Number'].str.replace(r'(\d{3})-\d{3}-(\d{4})', r'\1-****-\2', regex=True)
df_anon['SSN'] = df_anon['SSN'].str.replace(r'(\d{3})-\d{2}-(\d{4})', r'\1-**-\2', regex=True)


# Drop original Name column
df_anon.drop('Name', axis=1, inplace=True)
```

## Pseudonymization

With pseudonymization the transfered data can be encrypted before sending. If the receiver has the encryption key which was used to encrypt the data, then it can be decrypted to the original data.

```python
from cryptography.fernet import Fernet
```

```python
key = Fernet.generate_key()
cipher_suite = Fernet(key)

df_pseudo = df.copy()

df_pseudo['Email'] = df_pseudo['Email'].apply(lambda x: cipher_suite.encrypt(x.encode()))

display(df_pseudo)
```

Decrypt data to get the real values

```python
df_pseudo['Email'].apply(lambda x: cipher_suite.decrypt(x))
```
